```python
from ultralytics import YOLO
import cv2
```

```python
#model = YOLO("yolo11n.pt")  # load a pretrained model
model = YOLO("runs/train/yolo11n/weights/best.pt")
```

```python
results = model.train(data= "datasets/data.yaml",  # path to dataset
                      epochs=30,  # number of training epochs
                      batch=8,  # batch size
                      imgsz=416,  # image size
                      device=0,  # device to train on (0 for GPU, 'cpu' for CPU)
                      workers=4,  # number of data loading workers
                      project="runs1/train",  # directory to save results
                      name="yolo11nn",  # name of the experiment
                      #exist_ok=False,
 # device to train on (0 for GPU, 'cpu' for CPU
 )
```

```python
res = model(["8_aremicains/dataset/train/images/8.jpg"])
```

```python
# Process results list
for result in res:
    boxes = result.boxes  # Boxes object for bounding box outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs  # Probs object for classification outputs
    obb = result.obb  # Oriented boxes object for OBB outputs
    result.show()  # display to screen
```

```python
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Erreur : Impossible d'accéder à la webcam.")
    exit()
```

```python
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Effectuer la détection sur l'image de la webcam
    results = model(frame)

    # Afficher les résultats avec les bounding boxes
    annotated_frame = results[0].plot()

    # Affichage dans une fenêtre OpenCV
    cv2.imshow("Détection d'objets YOLOv8", annotated_frame)

    # Quitter avec la touche 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Libération des ressources
cap.release()
cv2.destroyAllWindows()
```
